```python
#data Preprocessing
```

```python
#Data Transformation
thyroid_df["binaryClass"].value_counts()
thyroid_df["binaryClass"]=thyroid_df["binaryClass"].map({"P":0,"N":1})
```

```python
thyroid_df["pregnant"].value_counts()
thyroid_df=thyroid_df.replace({"t":1,"f":0})
```

```python
thyroid_df=thyroid_df.replace({"F":1,"M":0})
```

```python
thyroid_df
```

```python
#convert object to corresponding data type
cols = thyroid_df.columns[thyroid_df.dtypes.eq('object')]
thyroid_df[cols] = thyroid_df[cols].apply(pd.to_numeric, errors='coerce')
thyroid_df.dtypes
```

```python
#find Missing Values
thyroid_df=thyroid_df.replace({"?":np.NAN})
thyroid_df.isnull().sum()
```

```python
#Filling Missing Values with Mean and remove unwanted attribute
thyroid_df['T4U measured'].fillna(thyroid_df['T4U measured'].mean(), inplace=True)
thyroid_df['sex'].fillna(thyroid_df['sex'].mean(), inplace=True)
thyroid_df['age'].fillna(thyroid_df['age'].mean(), inplace=True)

del thyroid_df["referral source"]
del thyroid_df["TBG"]

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')

thyroid_df['TSH'] = imputer.fit_transform(thyroid_df[['TSH']])
thyroid_df['T3'] = imputer.fit_transform(thyroid_df[['T3']])
thyroid_df['TT4'] = imputer.fit_transform(thyroid_df[['TT4']])
thyroid_df['T4U'] = imputer.fit_transform(thyroid_df[['T4U']])
thyroid_df['FTI'] = imputer.fit_transform(thyroid_df[['FTI']])

thyroid_df.isnull().sum()
```

Training Model

```python
#Taking Key Elements for Model Training
df = thyroid_df[["age", "sex","TSH","T3","TBG measured","T4U","TT4","FTI","binaryClass"]]
df
```

KNN

```python
# Importing and training the KNN Classifier model
from sklearn.model_selection import train_test_split
X = df.iloc[:, :-1].values
y = df.iloc[:, 8].values
x_train, x_test, y_train, y_test=train_test_split(X, y,test_size=0.2,random_state=3)
from sklearn.neighbors import KNeighborsClassifier  
knn=KNeighborsClassifier(n_neighbors=3, weights="distance", metric="euclidean")
knn.fit(x_train,y_train)

# Accuracy of the model
from sklearn.metrics import accuracy_score
y_pred=knn.predict(x_test)
knn_accuracy=accuracy_score(y_test, y_pred)*100
print("Accuracy of test set=",knn_accuracy)
```

Naive Bayes

```python
nb_df= thyroid_df[["TSH","T3","TT4","binaryClass"]]
X = nb_df.iloc[:, :-1].values
y = nb_df.iloc[:, 3].values
```

```python
#Since our dataset containing character variables we have to encode it using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,0] = le.fit_transform(X[:,0])

#Train test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)

#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#Training the Naive Bayes model on the training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)

#Predict the Result
y_pred  =  classifier.predict(X_test)
nb_accuracy=accuracy_score(y_test, y_pred)*100
print("Accuracy of Naive Bayes Classification =",nb_accuracy)
```

Random Forest Classification

```python
dt_df= thyroid_df[["TSH","T3","TT4","binaryClass"]]
X = dt_df.iloc[:, :-1].values
y = dt_df.iloc[:, 3].values
```

```python
# Splitting the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)


from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

# prediction on test set
y_pred=clf.predict(X_test)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
rf_accuracy=metrics.accuracy_score(y_test, y_pred)*100
print("Accuracy of Random Forest:",rf_accuracy)
```

```python
print(" Accuracy of Various Classification Algorithms: ")
print("Accuracy of KNN Algoirthm:",knn_accuracy)
print("Accuracy of Naive Bayes:",nb_accuracy)
print("Accuracy of Random Forest:",rf_accuracy)
```
